For the following arrays:

Prove following with Numpy:
1. Prove A.A−1 = I.

.

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

1) Prove A.A−1 = I.

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])


A_inv = np.linalg.inv(A)
I = A @ A_inv
print(I)
```

2. Prove AB ̸= BA.

```python
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB ≠ BA:", not np.array_equal(AB, BA))
```

3. Prove (AB)

T = BTAT
# .

```python
AB_T = AB.T
BT_AT = np.dot(B.T, A.T)
print("(AB)ᵀ = BᵀAᵀ:", np.array_equal(AB_T, BT_AT))
```

Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

```python
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])
B = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)

print(f"x = {X[0]}, y = {X[1]}, z = {X[2]}")
```
