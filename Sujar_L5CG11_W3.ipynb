Fare column only

```python
fare = titanic["Fare"]
fare.head()
```

Pclass and Age (in order)

```python
classAge = titanic[["Pclass","Age"]]
classAge.head()
```

Survived and Sex (in order)

```python
survivedAndSex = titanic[["Survived","Sex"]]
survivedAndSex.head()
```

Problem - 2 - Subsetting:

1. Filter the Titanic dataset for cases where the passengers fare is greater than 100, assigning it
to fare gt 100.View the printed result.

```python
fare = titanic[titanic["Fare"]>100]
fare.head()
```

First-class passengers

```python
firstClass = titanic[titanic["Pclass"] ==1]
firstClass.head()
```

Female passengers under 18

```python
femaleAge = titanic[
    (titanic["Age"] < 18) & (titanic["Sex"] == "female")
]
femaleAge.head()
```

Subsetting by Categorical Variables

1. Embarked from C or S

```python
embarked_c_or_s = titanic[titanic["Embarked"].isin(["C", "S"])]
embarked_c_or_s.head()
```

First or second class passengers

```python
first_second_class = titanic[titanic["Pclass"].isin([1, 2])]
first_second_class.head()
```
